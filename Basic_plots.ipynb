# Plotting simple functions

- one-one line
- sin(x) & cos(x)
- multiple subplots
- different line clours
- different line styles

If we don't define a marker <b>matplotlib.pyplot.plot()</b> will draw a line. 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 

#generate some data to plot:
x = np.arange(0,20,0.1)

# plot the data against itself:
fig, ax = plt.subplots()
ax.plot(x, x)
plt.show()
```

```python
x = np.arange(0,20,0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'blue')
ax.plot(x, z, 'red')
plt.show()
```

Let's draw 2 subplots, one with sin(x) and one with cos(x).

We will also define a custom the figure size and change the linestyle.

More about line properties here: https://matplotlib.org/2.0.2/api/lines_api.html

```python
x = np.arange(0,20,0.1)
y = np.sin(x)
z = np.cos(x)
a = np.sin(x+5)
b = np.cos(x+5)

fig = plt.figure(figsize=(10,4))  #custom figure size
ax = fig.add_subplot(1,2,1)  # to create multiple subplots
ax.plot(x, y, 'C0')
ax.plot(x, a, 'C3', linestyle=':')
plt.title('Sin(x)')
plt.axvline(10, color='grey', linestyle = '--')

ax = fig.add_subplot(1,2,2)
ax.plot(x, z, 'C2', linestyle = '--')
ax.plot(x, b, 'k', linestyle='-.')
plt.title('Cos(x)')
plt.show()
```

# Scatter plots

A very simple way to make a scatter plot is using <b>matplotlib.pyplot.plot()</b>.

<b>Resources:</b>

https://matplotlib.org/gallery/api/unicode_minus.html#sphx-glr-gallery-api-unicode-minus-py


```python
# Create some random numbers - Fixing random state for reproducibility
np.random.seed(19680801)
x = 10*np.random.randn(100)
y = 10*np.random.randn(100)

fig, ax = plt.subplots()
ax.plot(x, y, 'd')

# to save the plot as an individual pdf file:
#plt.savefig('scatter.pdf')  

plt.show()
```

## Changing the properties of the plot

Let us change the following things:
    - marker colour
    - marker size
    - marker shape
    - add a title to the plot
    - add lables to the axis
    - draw some vertical and horizontal lines
    - write some text onto the figure
    
    
Information on matplotlib markers: https://matplotlib.org/api/markers_api.html

Information on matplotlib default colours: https://matplotlib.org/users/dflt_style_changes.html

```python
fig, ax = plt.subplots()
ax.plot(x, y, '>', mfc = 'r', mec = 'k', markersize = 10)  # mec - marker edge clour, mfc - marker face colour
 
plt.xlabel('Random numbers')
plt.ylabel('Random numbers')
plt.title('Scatter plot')

#let's add some horizontal and vertical lines
plt.axhline(-30, color='grey')
plt.axvline(-30, color='grey', linestyle = '--')

#let's add some text
plt.text(-25,-25, 'random', fontsize=16)

#plt.show()
plt.savefig('test.png')
```

Another way of drawing scatter plots is using <b>matplotlib.pyplot.scatter()</b>. This way we have more options to change the properties of the markers. E.g. different sized and cloured markesr.

<b>Resources:</b> 

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html

https://matplotlib.org/gallery/lines_bars_and_markers/scatter_symbol.html#sphx-glr-gallery-lines-bars-and-markers-scatter-symbol-py

Information on matplotlib colourmaps: https://matplotlib.org/examples/color/colormaps_reference.html


```python
np.random.seed(19680801)
x = 10*np.random.randn(100)
y = 10*np.random.randn(100)
s = np.arange(len(x)) * 2

plt.scatter(x, y, s=s, c=s, cmap="jet", alpha=0.5, marker='s')
 
plt.xlabel('Random numbers')
plt.ylabel('Random numbers')
plt.title('Scatter plot')

plt.show()
```

Let's change up a few things in our scatter plot.

- we can add a colourbar with <b>matplotlib.pyplot.figure.colorbar()</b>
- we can add a legend using <b>label=</b> and <b>matplotlib.pyplot.legend()</b>
- markers can also be latex symbols

```python
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
s = np.random.rand(*x.shape) * 800 + 500

fig = plt.figure()
luck = plt.scatter(x, y, s=s, c=s, cmap="viridis", alpha=0.5, marker=r'$\clubsuit$', label="Luck")
plt.xlabel("Leprechauns")
plt.ylabel("Gold")
plt.legend(loc='upper left', fontsize=14, numpoints=1) # add a legend
fig.colorbar(luck) # add a colourbar 
plt.show()
```

<b>Some resources:</b>

https://matplotlib.org/gallery/statistics/hist.html#sphx-glr-gallery-statistics-hist-py

One of the simplest ways to plot histograms is to use: <b>matplotlib.pyplot.hist()</b>

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, bins = 50,  histtype='bar', facecolor='g', alpha=0.75)    
    
plt.xlabel('Smarts')
plt.ylabel('N')
plt.title('Histogram')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)  # this controlls the grid for the figure 
plt.show()
#plt.savefig('hist.pdf')
```

Let us modify the style of the histogram a bit.
    - change the histogram style to step
    - change the colour to red
    - make the grid dashed
    - change the font size for the title and axis labels
    - change the font style for the title
    - change the spacing of the axis ticks to 10
    - change the size of the yxis ticks

```python
# the histogram of the data

fig = plt.figure()
ax = fig.add_subplot()

plt.hist(x, bins = 50, histtype='step', color='r', hatch='\\', label='hist')

# add labels
plt.xlabel('Smarts', fontsize=16)
plt.ylabel('N', fontsize=16)
plt.title('Histogram', fontsize=18, family='serif')

# change grid properties:
plt.grid(True, linestyle='--')  
plt.xticks(np.arange(40,170,10), size=10)

# change ticks
plt.yticks(size=13)

# set axis boundaries:
plt.xlim(60, 180)
plt.ylim(0, 800)

# add a legend
plt.legend(loc='upper right', fontsize=18)
    
plt.show()
```

## Multiple histograms on one plot

```python
#Make the histogram
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
specis = []


# open the file, read the lines, and split the columns into lists
g = open('./data/iris.csv', 'r')
data = g.readlines()
for j in range(1, len(data)):
    c = data[j].split(',')
    sepal_length.append(float(c[0]))  #float() because we know that this column only has numbers
    sepal_width.append(float(c[1]))
    petal_length.append(float(c[2]))
    petal_width.append(float(c[3]))
    specis.append(c[4])
    
g.close()


fig, ax = plt.subplots()

bins = np.arange(0,9,0.25)

plt.hist(petal_length, bins=bins, histtype='step', color='C0', lw=2,  label = 'petal length')
plt.hist(petal_width, bins=bins, histtype='step', color='C1',  lw=2, label = 'petal width', hatch = '\\')
plt.hist(sepal_length, bins=bins, histtype='step', color='C2', lw=2, label = 'sepal length')
plt.hist(sepal_width, bins=bins, histtype='step', color='C3', lw=2, label = 'sepal width', hatch = '//')

plt.xlabel('Flower properties', fontsize=18, family='serif')
plt.ylabel('N', fontsize=18, family='serif')
plt.title('Iris Histogram', fontsize=18, family='serif')
plt.legend()
plt.show()
```

# Bonus - How to draw a hopalong fractal

Based on this blogpost:

https://glowingpython.blogspot.com/2013/07/hopalong-fractals.html

```python
from numpy import sqrt,power
import pylab as P 
from pylab import scatter,show, cm, axis
from numpy import array,mean
#---------------------------------------
# plot hopalong fractal
# change a, b, c  to change fractal pattern
#---------------------------------------

def hopalong(x0,y0,n,a,b,c):
    def update(x,y):
        x1 = y-x/abs(x)*sqrt(abs(b*x+c))
        y1 = a-x
        return x1,y1
    xx = []
    yy = []
    for x in range(n):
        x0,y0 = update(x0,y0) 
        xx.append(x0)
        yy.append(y0)
    return xx,yy


fig = P.figure()
x = -1
y = 10
n = 40000
#xx,yy = hopalong(x,y,n,1,5,5)
xx,yy = hopalong(x,y,n,-55,-1,-42)
cr = sqrt(power(array(xx)-mean(xx),2)+power(array(yy)-mean(yy),2))
P.scatter(xx, yy, marker='.', c=cr/max(cr), edgecolor='w', cmap=cm.viridis_r, s=50)
axis('equal')
show()
#P.savefig('fractal_2.png')
```
