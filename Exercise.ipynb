# Exercise 1

Read in data table (iris.csv) into lists. One list for each column (sepal_length, sepal_width, petal_length, petal_width, specis)

1. Define lists
2. open the file
3. read in the individual lines in the file
4. split up the lines based on the deliminator (',')
5. add data to lists
6. print your lists to check them


```python
#Read in data from a table (iris.csv)

# define the lists in which you read in the data
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
specis = []


# open the file, read the lines, and split the columns into lists
g = open('./data/iris.csv', 'r')
data = g.readlines()
for j in range(1, len(data)):
    c = data[j].split(',')
    sepal_length.append(float(c[0]))  #float() because we know that this column only has numbers
    sepal_width.append(float(c[1]))
    petal_length.append(float(c[2]))
    petal_width.append(float(c[3]))
    specis.append(c[4])
    
g.close()

print(sepal_length) 
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.arange(-20,20,0.1)
y = np.sin(np.pi * x)/(np.pi * x)

fig, ax = plt.subplots()
ax.plot(x, y, 'C3')
ax.axhline(0, color='grey', linestyle='--')
plt.show()
```

```python
#Read in data from a table (iris.csv)

# define the lists in which you read in the data
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
specis = []


# open the file, read the lines, and split the columns into lists
g = open('./data/iris.csv', 'r')
data = g.read().splitlines()
for j in range(1, len(data)):
    c = data[j].split(',')
    sepal_length.append(float(c[0]))  #float() because we know that this column only has numbers
    sepal_width.append(float(c[1]))
    petal_length.append(float(c[2]))
    petal_width.append(float(c[3]))
    specis.append(c[4])
    
g.close()
```

```python
plt.scatter(petal_length, petal_width, s=15, c='green', alpha=0.5, marker='s', label = 'petals')
#plt.plot(petal_length, petal_width, 's', markersize=5, mfc='green', mec='green', alpha=0.5 , label = 'petals')
 
plt.xlabel('Petal length', fontsize=16)
plt.ylabel('Petal width', fontsize=16)
plt.title('Iris petals', fontsize=16)
plt.legend()

plt.show()
```

```python
#Make the histogram

plt.hist(petal_length, histtype='step', color='r', fill=True, label = 'petals')

plt.xlabel('Petal lengths', fontsize=18, family='serif')
plt.ylabel('N', fontsize=18, family='serif')
plt.title('Petal Length Histogram', fontsize=18, family='serif')
plt.grid(True, linestyle=':', color='C0')  # this controlls the grid for the figure 
plt.legend()
plt.show()
```
